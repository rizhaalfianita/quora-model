```python
import pickle

with open('./60-40/performance_score_40.pkl', 'rb') as f:
    performance_score_40 = pickle.load(f)

with open('./70-30/performance_score_30.pkl', 'rb') as f:
    performance_score_30 = pickle.load(f)

with open('./80-20/performance_score_20.pkl', 'rb') as f:
    performance_score_20 = pickle.load(f)

with open('./90-10/performance_score_10.pkl', 'rb') as f:
    performance_score_10 = pickle.load(f)
```

```python
scenarios = ['Skenario 1', 'Skenario 2', 'Skenario 3', 'Skenario 4']
accuracy_values = [performance_score_40[0], performance_score_30[0], performance_score_20[0], performance_score_10[0]]
precision_values = [performance_score_40[1], performance_score_30[1], performance_score_20[1], performance_score_10[1]]
recall_values = [performance_score_40[2], performance_score_30[2], performance_score_20[2], performance_score_10[2]]
f1_values = [performance_score_40[3], performance_score_30[3], performance_score_20[3], performance_score_10[3]]
```

```python
print(accuracy_values)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Define custom color palette with shades of red and reverse it
custom_palette = sns.color_palette("Reds_r", len(scenarios))[::-1]

sns.set_style('whitegrid')
ax = sns.barplot(x=scenarios, y=accuracy_values, palette=custom_palette)
plt.title('Comparison of Accuracy', fontsize=10, fontweight='bold', pad=11)
plt.xlabel('Scenarios', fontsize=9, fontweight='semibold', labelpad=6)  # Add x-axis title
plt.ylabel('Accuracy', fontsize=9, fontweight='semibold', labelpad=6)   # Add y-axis title
plt.ylim(0, 1)

# Add numeric values above the bars
for i, v in enumerate(accuracy_values):
    ax.text(i, v + 0.01, str(round(v, 2)), ha='center', va='bottom', fontsize=8)

plt.show()
```
